```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture("/Users/ACER/Desktop/Machine Vision/Udacity - Self-Driving Car Engineer - A/Project 1 - Finding Lane Lines/test_videos/solidYellowLeft.mp4")
#cap = cv2.VideoCapture(0)
```

```python
rho = 1
theta = np.pi/180
threshold = 50
minLineLength =50
maxLineGap =400
```

```python
left_x = []
left_y = []
right_x = []
right_y = []

top =350 
bottom = 540

#op=480 #or challenge
#ottom = 630
```

```python
def drawLine(x1,y1,x2,y2,frame):
    
    slope = (y2-y1)/(x2-x1)
    print(slope)
    if (slope >0.3):
        
        left_x.append(x1)
        left_x.append(x2)
        left_y.append(y1)
        left_y.append(y2)
        
    if (slope <-0.3):
        
        right_x.append(x1)
        right_x.append(x2)
        right_y.append(y1)
        right_y.append(y2)
        
        
    if (left_x !=[]) & (left_y !=[]) & (right_x !=[]) & (right_y !=[]) :
        leftFunc = np.polyfit(left_x,left_y,1)
        left_x_top = int((top-leftFunc[1])/leftFunc[0])
        left_x_bottom = int((bottom-leftFunc[1])/leftFunc[0])
        
        rightFunc = np.polyfit(right_x,right_y,1)
        right_x_top = int((top-rightFunc[1])/rightFunc[0])
        right_x_bottom = int((bottom-rightFunc[1])/rightFunc[0])
        
        cv2.line(oriFrame,(left_x_top,top),(left_x_bottom,bottom),(0,0,255),10)
        cv2.line(oriFrame,(right_x_top,top),(right_x_bottom,bottom),(0,0,255),10)
        
        
        
        
        
```

```python
while True:
    ret,frame = cap.read()
    
    if ret == True:
        oriFrame = np.copy(frame)
        copyFrame = np.copy(frame)
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        frame = cv2.GaussianBlur(frame,(5,5),0)
        frameShape = frame.shape
        #threePoint = np.array([[(590,450),(690,450),(1280,660),(0,660)]],dtype = np.int32)
        threePoint = np.array([[(430,350),(530,350),(960,540),(0,540)]],dtype = np.int32)
        canny = cv2.Canny(frame,50,300)
        blackBG = np.zeros_like(canny)
        mask = cv2.fillPoly(blackBG,threePoint,(255,255,255))
        filteredCanny = cv2.bitwise_and(mask,canny)
        cannys = np.dstack((canny,canny,canny))
        lines = cv2.HoughLinesP(filteredCanny,rho,theta,threshold,np.array([]),minLineLength,maxLineGap)
        try:
            for line in lines:
                for x_1,y_1,x_2,y_2 in line:
                    #cv2.line(oriFrame,(x1,y1),(x2,y2),(0,0,255),10)
                    #slope = (y2-y1)/(x2-x1)
                    drawLine(x_1,y_1,x_2,y_2,oriFrame)
        except:
            pass
                    
                    
        showFrame = cv2.addWeighted(oriFrame,0.5,copyFrame,0.5,0)
        cv2.imshow('road',showFrame)
    
    if cv2.waitKey(25) == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
